# üõí Group A ‚Äì Smart Grocery Store App

## üìù Problem Statement

You are asked to build a console-based Grocery Store App.  
The program should allow a customer to:

- **Show available items** (stored in a dictionary of items and prices).
- **Add an item to the cart** (store selected items in a list or set).
- **Remove an item from the cart**.
- **View the cart** (list all items and total cost).
- **Checkout**:
  - Calculate the total price using a function.
  - Apply a 10% discount if total > 20.
- **Exit** the program when the user chooses.

---

### üîπ Example Store Items
```python
store = {
    "apple": 2,
    "banana": 1,

}

````python

store = {
    "apple": 2,
    "banana": 1,
    "milk": 3,
    "bread": 2,
    "eggs": 5
}

cart = {}

price ={
    "apple": 20,
    "banana": 10,
    "milk": 30,
    "bread": 20,
    "eggs": 50
    }

final_amontlst={}

def avlble_items():
    print(store)
    

def additem_cart():
    new_item = str(input("Add new item to the cart"))
    number = int(input("Enter the number of item"))
    if new_item in store:
        if store[new_item] >= number:
            store[new_item]-=number                
            if new_item in cart:        
                cart[new_item]+=number
            else:
                cart[new_item] = number 
        else:
            print(f"Please enter the number lower than {number}, not enough stock")
                 
    else:
        print("Item is not available")
    print("Your item is successfully added to cart")


def removeitem():
    wantto_remove = str(input("Enter the item to remove"))
    if wantto_remove in cart:
        cart.pop(wantto_remove)
        print(cart)
    else:
        print("Item not in cart")
            
def calculate_total():
    total = 0
    final_amontlst.clear()
    for item, qty in cart.items():
        final_amontlst[item] = qty * price[item]
    total = sum(final_amontlst.values())
    return total

def view_cart():
    total = calculate_total()
    print("Your cart is",cart)
    print("Total amount",total)

def checkout():
    total = calculate_total()
    if total > 20:
        discount = total * 0.10
        Discounted_price = total - discount
        print("You received a 10% discount and the Discounted Price",Discounted_price)
    else:
        print("Final amount to pay",total)
    print("‚úÖ Thank you for shopping with us!")
    cart.clear()
    

while True:
    print("Grocery Store")
    print("1. Show Available Items")
    print("2. Add an item")
    print("3. Remove an item")
    print("4. View the cart")
    print("5. Checkout")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == "1":
        avlble_items()
    elif choice == "2":
        additem_cart()
    elif choice == "3":
        removeitem()
    elif choice == "4":
        view_cart()
    elif choice == "5":
        checkout()
    elif choice == "6":
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
````

# üìö Group B ‚Äì School Library App

## üìù Problem Statement

You are asked to build a console-based Library System.  
The program should allow a student to:

- **Show available books** (stored in a dictionary of books and available copies).
- **Borrow a book** (store borrowed books in a list or set, and reduce available copies).
- **Return a book** (increase available copies).
- **View borrowed books**.
- **Checkout**:
  - Calculate the total fee using a function.
  - Apply a $2 late fee if more than 3 books are borrowed.
- **Exit** the program when the user chooses.

---

### üîπ Example Library Books
```python
library = {

}

````python
sample = {
    "Python Basics": 3,
    "Data Science 101": 2,
    "AI for Beginners": 4,
    "Machine Learning": 1
}

Avlb_books = sample.copy()
borrow_book = []
def available_book():
    print("Available Books")
    for book,copies in Avlb_books.items():
        Avlb_books[book] = copies
    print(Avlb_books)



def borrow():
    book_need = str(input("Enter the book you want to borrow"))
    if book_need in Avlb_books and Avlb_books[book_need]>0:
        borrow_book.append(book_need)
        Avlb_books[book_need] = Avlb_books[book_need] -1
    else:
        print("Book not available")
    #available_book()
    print("You borrowed these books",borrow_book)

def return_book():
    book_return = str(input("Enter the book to return"))
    if book_return in Avlb_books:
        Avlb_books[book_return] = Avlb_books[book_return] +1
        borrow_book.remove(book_return)
        print(f"you returned {book_return}")
    else:
        print("You didn't borrow that book")
    available_book()



def view_borrowed():
    print("View Borrowed Books",borrow_book)

def check_out():
    print("Checkout Summary")
    total_books = len(borrow_book)
    print(f"Total books borrowed: {total_books}")
    late_fee = 2 if total_books > 3 else 0
    print(f"Late Fee: ${late_fee}")


while True:
    print("School Library Menu")
    print("1. Show Available Books")
    print("2. Borrow a Book")
    print("3. Return a Book")
    print("4. View Borrowed Books")
    print("5. Checkout")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == "1":
        available_book()
    elif choice == "2":
        borrow()
    elif choice == "3":
        return_book()
    elif choice == "4":
        view_borrowed()
    elif choice == "5":
        check_out()
    elif choice == "6":
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")

````
