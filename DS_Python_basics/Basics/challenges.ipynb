## **Group 1 Challenge ‚Äì "Classroom Candy Distribution" üç¨**

**Scenario:**  
You‚Äôre helping a teacher distribute candies to students in a fair way.

### Instructions:
1. Start with the following list of student names (some names are duplicated):
   ```python
   students = ["Ali", "Sara", "Ali", "John", "Sara", "Maya", "Zain", "Maya", "Omar"]

````python
students = ["Ali", "Sara", "Ali", "John", "Sara", "Maya", "Zain", "Maya", "Omar"]
````

2. Remove duplicates using a set, then convert back to a list:
   ```python

````python
students = ["Ali", "Sara", "Ali", "John", "Sara", "Maya", "Zain", "Maya", "Omar"]
students = set(students)
students1 = list(students)
print(students1)
````

3. Each student gets candies based on the length of their name:

- Name length ‚â§ 4 ‚Üí 2 candies
- Name length 5‚Äì6 ‚Üí 3 candies
- Name length > 6 ‚Üí 4 candies

````python
candies = {}
for i in students1:
    if len(i) <= 4:
        candies[i] =2
    elif len(i) == 5 or len(i) == 6:
        candies[i] =3
    else:
        candies[i] =4
print(candies)
````

4. Store the result in a list of tuples in the format:

[("Ali", 2), ("Sara", 3), ("John", 3), ...]

````python
tuple1 = tuple(candies.items())
print(tuple1)
````

5. Calculate total candies needed.


````python
sum =0
dict_1 = dict(tuple1)
for i in dict_1.values():
    sum += i
print(sum)
````


6. Print the results in this format:

Ali gets 2 candies.<br>
Sara gets 3 candies.<br>
...<br>
Total candies needed: 25<br>

````python
for x, i in dict_1.items():
    print(f"{x} gets {i} candies")
````

# üèÜ Group 2 Challenge ‚Äì "Sports Tournament Scoring"

**Scenario:**  
You‚Äôre in charge of calculating scores for players in a tournament.

---

## **Instructions**
1. Start with the following list of player names (some names are duplicated):
   ```python
   players = ["Lina", "Omar", "Lina", "Hadi", "Omar", "Noor", "Hadi", "Karim"]
   

````python
players = ["Lina", "Omar", "Lina", "Hadi", "Omar", "Noor", "Hadi", "Karim"]
````

2. Remove duplicates using a set, then convert back to a list.

````python
players=set(players)
print(players)
players=list(players)
print(players)
````

3. Each player‚Äôs starting score is based on the number of letters in their name:

Name length ‚â§ 4 ‚Üí 5 points <br>
Name length 5‚Äì6 ‚Üí 10 points <br>
Name length > 6 ‚Üí 15 points <br>

````python
points= {}
for i in players:
    if len(i) <= 4:
        points[i] = 5 
    elif len(i) == 5 or len(i) ==6:
        points[i] = 10
    else:
        points[i] =15
print(points)
````

4. Store the result in a list of tuples in the format:

[("Lina", 5), ("Omar", 5), ("Hadi", 5), ...]


````python
tuple1 = tuple(points.items())
print(tuple1)
````

5. Calculate total points across all players.

````python
sum = 0
dict1 = dict(tuple1)
for i in dict1.values():
    sum+=i
print(sum)
````


6. Print the results in this format:

Lina starts with 5 points. <br>
Omar starts with 5 points. <br>
...<br>
Total points: 50<br>


````python
for x,i in dict1.items():
    print(f"{x} starts with {i} points")
````
