### Cell types

You might have started to notice that the code blocks have little notes that say In **[ ]**: before you run them, and then are filled with a  number after <br> you run them. This is important, as it tells you what order cell blocks were run. (Which can sometimes affect how a program runs.)

You may also notice that other cell blocks, such as this one, do not have the In **[ ]**: label as with the code blocks. This is because this <br> cell block is formatted as Markdown (rather then code. The details of Markdown are not important here but just know you can use <br> markdown cell blocks to display text. They are really useful for adding notes and explanations in your Jupyter Notebook. You can <br> see (and change) what type of cell is by clicking the dropdown menu at the top.

## Command vs edit mode

We distinguish between two modes in jupyter lab, one is the edit mode where you can write code or notes, the other is the command <br> mode where you can delete or copy cells.

To switch from edit mode to command mode, you can press **esc**, (you cannot write in the cell now) and press Enter to **enter** edit mode <br> again.

### Edit mode
Edit mode is the standard mode for editing cells, whether it's writing code or notes. To enter edit mode from command mode simply hit <br> enter, or double click on a cell.

### Command mode
In command mode, you can delete cells, add cells, copy cells, paste cells, change cell types, and more. You can also do these tasks in <br> a more time consuming manner by using the various headers in the menu bar at top.
If you want to learn more about the<br> functionalities of jupyter lab, here is the here is the [documentation](https://jupyterlab.readthedocs.io/en/latest/).
