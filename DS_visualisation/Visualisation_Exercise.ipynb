3) Use lineplot from Matplotlib to understand the relation between mpg and weight.
- Make sure you sort the dataset by the 'weight' column, before drawing a line plot
- Add titles and labels

````python
df_weight = df.sort_values('weight',ascending=True)
plt.plot(df_weight['mpg'], df_weight['weight'],label='MPG vs Weight')
plt.title('Relation between mpg and weight')
plt.xlabel('MPG')
plt.ylabel('Weight')
plt.legend()
plt.show()

#As weight decreases, MPG tends to increase and lighter cars usually have better fuel efficiency
````

4) A scatter plot helps us understand the relationship between two numeric variables. Use seaborn to plot a scatterplot showing the relationship between<br> weight and Acceleration of the car.
- Add labels
- Interpret the result

````python
sns.scatterplot(x='acceleration', y='weight', data=df,hue='acceleration')
plt.title('Relation between Acceleration and weight')
plt.xlabel('Acceleration')
plt.ylabel('Weight')
plt.show()

#As acceleration increases, weight tends to decrease. This makes sense — lighter cars usually accelerate faster.
#But there are fewer cars with very high acceleration and very high weight (top right).
````

5) Histograms provide a visual representation of the distribution of a numeric variable. Use the library Plotly to draw a histogram for the column `HorsePower`<br>
- Add title, labels 
- What does the result say?
- How does Plotly different from other two libraries?

````python
import plotly.express as px

fig = px.histogram(df, x="horsepower",color="origin",title="Distribution of Horsepower",labels={"horsepower": "Horsepower (HP)"}) 
fig.show()

#From the plot, you can see:

# Most cars have horsepower concentrated around 50-100 HP.
# Cars from origin "1" have the highest counts(no.of cars).
# Cars from origins "2" and "3" have fewer cars and are scattered more in lower or higher horsepower ranges.

#plotly plots are interactive by default — you can zoom, pan, hover for details                 
````

We can visualize categorical data using bar plots, pie charts, and box plots.

6) Use boxplot from Seaborn to understand the Cars Distribution by Weight for each Origin value
- what do you see? Explain!

````python

sns.boxplot(x='origin', y='weight', data=df,hue='origin')
plt.title('Relation between origin and weight')
plt.xlabel('origin')
plt.ylabel('Weight')
plt.show()

#Cars from origin 1 generally have the highest weights
#Cars from origin 2 have a lower median weight
#Cars from origin 3 have the lowest weights
````

7) Plot a similar boxplot to understand the Cars Distribution by Acceleration for each Origin values
- what do you see? Explain.

````python
plt.figure(figsize=(8,6))
sns.boxplot(x='origin', y='acceleration', data=df,hue='origin')
plt.title('Relation between origin and acceleration')
plt.xlabel('origin')
plt.ylabel('acceleration')
plt.show()

#Cars from origin 1 have the lowest median acceleration
#Cars from origin 2 have the highest median acceleration
#Cars from origin 3 have a median acceleration
````
