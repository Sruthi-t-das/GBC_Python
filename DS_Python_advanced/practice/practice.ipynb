2- using seaborn use the lineplot to plot the Rings and height of abalone  (Rings on x axis) with the following specifications:

	- dark style

	- figure size = 20,5

	- without spines

```python
# draw lineplot 

sns.set_style("dark")
plt.figure(figsize = (20, 5))
sns.lineplot(x="Rings", y="Height", data=df) 
sns.despine()
plt.show()
```

```python
#3- write a function that can plot a lineplot for the whole weight and the diameter of the abalone with temporarily whitegrid style

def drawline():
    sns.lineplot(x="Whole weight", y="Diameter", data=df)
    plt.show() 
with sns.axes_style('whitegrid'):
      drawline()
```

```python
# importing packages
import pandas as pd 
import matplotlib.pyplot as plt  
import seaborn as sns 
  
# loading dataset 

# df = pd.read_csv(r"C:\Users\sruth\Documents\Greenbootcamps\DS_Python_advanced\data\abalone.csv")
# df.head()
data = sns.load_dataset('titanic') 
data
```

1.	Distribution of Age
A.	Plot a histogram (sns.histplot) of passengers ages to explore the age distribution.
B.	Place it in the top-left of a 22 grid.


```python
# def graph():
# 	sns.histplot(x="age", data=data)
    # plt.show()
# graph()
fig, axs = plt.subplots(2, 2, figsize=(5, 3))
sns.histplot(x="age", data=data,ax=axs[0,0])
axs[0,0].set_title("Age Distribution")
plt.tight_layout()
plt.show()

```

2.	Survival Count by Gender
A.	Create a barplot (sns.countplot) showing the number of survivors vs non-survivors, separated by gender.
B.	Place it in the top-right of a 22 grid.


```python

fig, axs = plt.subplots(2, 2, figsize=(5, 3))
sns.countplot(x='survived', data=data,hue='sex',ax=axs[0,1])
axs[0,1].set_title("Survived - Gender")
plt.show()
```

3.	Class vs Fare
A.	Use a boxplot (sns.boxplot) to compare the distribution of fares across passenger classes (Pclass).
B.	Place it in the bottom-left of a 22 grid.


```python

fig, axs = plt.subplots(2, 2, figsize=(8, 6))
sns.boxplot(x='pclass',y ='fare', data=data,ax=axs[1,0])
axs[1,0].set_title("Survived - Gender")
plt.show()
```

6.	Fare Distribution by Survival (Boxplot)
A.	Draw a boxplot (sns.boxplot) comparing fare distribution between survivors and non-survivors.

7.	Histogram of Passenger Fare
A.	Plot a histogram (sns.histplot) of passenger fares, setting bins=30.

A.	Create a barplot (sns.countplot) showing survival counts across passenger classes (Pclass).

```python

plt.figure(figsize=(12, 10))  
axs = plt.subplot2grid((2, 2), (1, 0), colspan=2)
sns.countplot(x='pclass', data=data,hue='survived',ax=axs)
plt.tight_layout()
plt.show()
```
