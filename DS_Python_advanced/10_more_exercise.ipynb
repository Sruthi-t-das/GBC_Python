3. I've decided that this month I want to stay away from wines with relatively high alcohol content. To do that, I'm going<br> to avoid any wines that have a greater alcohol content than the mean alcohol content, and you're going to help me do this.<br> To achieve this, let's do the following:

- Find the mean alcohol content, separately, for reds and whites.
- Create a ```Series``` that holds whether each row in each ```DataFrame``` (the reds, and whites) has a higher alcohol content than<br> the mean.
- Merge this ```Series``` onto the ```DataFrame```. I can imagine doing this with either a ```.join()``` or using ```pd.concat()```. For practice, do it with both. Note: merges with ```Series``` work the same way that they work with ```DataFrames```.
- Return back to me all those rows that will help me stay away from those wines with a higher alcohol content.

````python
alcohol_series_red = red_wines_df.alcohol > red_wines_df.alcohol.mean()
alcohol_series_white = white_wines_df.alcohol > white_wines_df.alcohol.mean()
````

````python
# alcohol_series_red.rename('alcohol_high_red',inplace=True)
alcohol_series_red.name = 'alcohol_high_red'
````

````python
alcohol_series_white.rename('alcohol_high_white',inplace=True)
````

````python
# high_alcohol= pd.concat([red_wines_df, alcohol_series_red], axis=1)
# high_alcohol
Red_flag_Red = red_wines_df.join(alcohol_series_red)
Red_flag_Red
````

````python
Red_flag_Red[Red_flag_Red['alcohol_high_red'] == True]
````

````python
# high_alcohol= pd.concat([white_wines_df, alcohol_series_white], axis=1)
# high_alcohol
Red_flag_White = white_wines_df.join(alcohol_series_white)
Red_flag_White
````

````python
Red_flag_White[Red_flag_White['alcohol_high_white'] == True]
````

4. Let's say that I want to get started on cutting back next month. This time, though, I want to focus on staying away<br> from those wines with a high acidity. Specifically, I want to stay away from those wines that are in the highest bin of<br> fixed acidity (highest bin out of 5). You're now going to help me do this. To achieve this, let's do the following:

- Separate the rows in each DataFrame into 5 equal width bins (not equal to quintiles) based off their fixed acidity.
- Merge the resulting Series holding these 5 bins onto the original DataFrame. I can imagine also doing this with either . <br> 
    - join() or using pd.concat(). Try doing it with both for practice.
- Return back to me all those rows that are not in the top bin in terms of fixed acidity.

````python

fixed_acidity_red = pd.cut(red_wines_df['fixed acidity'], bins=5 ,labels=[1,2,3,4,5])
fixed_acidity_red.name = 'fa_bin'
fixed_acidity_red
FA_Red = Red_flag_Red.join(fixed_acidity_red)
FA_Red

# fixed_acidity_bins = np.arange(1, 7)
# fixed_acidity_red = pd.cut(red_wines_df['fixed acidity'], bins=fixed_acidity_bins,labels=fixed_acidity_bins[:-1])
# fixed_acidity_red.name = 'fa_bin'
# fixed_acidity_red
# FA_Red = Red_flag_Red.join(fixed_acidity_red)
# FA_Red

# fixed_acidity_series.name = 'fa_bin'
# red_wines_df = pd.concat([red_wines_df, fixed_acidity_series], axis=1)
````

````python
FA_Red[FA_Red['fa_bin'] != 5]
````

````python
fixed_acidity_white = pd.cut(white_wines_df['fixed acidity'], bins=5 ,labels=[1,2,3,4,5])
fixed_acidity_white.name = 'fa_binW'
fixed_acidity_white
FA_White = Red_flag_White.join(fixed_acidity_white)
FA_White
````

````python
FA_White[FA_White['fa_binW'] != 5]
````

5. Let's say that I now want to know how much my decision to avoid those wines with higher ```alcohol``` content is going to<br> limit the ```quality``` of wines that I can drink. To figure this out, I want to know a couple of things:


- The average ```alcohol``` content for those reds above the mean ```alcohol``` level, by quality.
- The average ```alcohol``` content for those whites above the mean ```alcohol``` level, by quality.
- Use a ```pivot table``` to solve this.

````python
pd.pivot_table(Red_flag_Red, values='alcohol', index='quality', columns='alcohol_high_red')
````

````python
pd.pivot_table(Red_flag_White, values='alcohol', index='quality', columns='alcohol_high_white')
````

6. Now, do the same for my decision to avoid wines with a high acidity next month:

- Find the average ```alcohol``` content for reds, by ```quality``` and ```fixed acidity``` bin.
- Find the average ```alcohol``` content for whites, by ```quality``` and ```fixed acidity``` bin.

````python
pd.pivot_table(FA_Red, values='alcohol', index='quality', columns='fa_bin',aggfunc=np.mean)
````

````python
pd.pivot_table(FA_White, values='alcohol', index='quality', columns='fa_binW',aggfunc=np.mean)
````
