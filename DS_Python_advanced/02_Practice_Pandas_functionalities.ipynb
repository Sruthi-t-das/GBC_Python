### Part 3 : More into the data

Let's move on to some more complicated things. Use your knowledge of ```groupby```s, ```sorting``` to answer the following.

1. Get the average amount of ```chlorides``` for each ```quality``` value.

````python
df.head(5)
````

````python

# df.groupby('quality')['chlorides'].mean()
df.groupby('quality').mean()['chlorides']
````

2. For observations with a ```pH``` greater than 3.0 and less than 4.0, find the average ```alcohol``` value by ```pH```.

````python
df[(df['pH']>3.0) & (df['pH']<4.0)].groupby('pH')['alcohol'].mean()
df.query("pH>3.0 and pH<4.0").groupby('pH')['alcohol'].mean()
````

3. For observations with an ```alcohol``` value between 9.25 and 9.5, find the highest amount of ```residual sugar```.

````python
df[(df['alcohol']>9.25) & (df['alcohol']<9.5)].sort_values('residual sugar',ascending=False).head()
````

4. Create a new column, called ```total_acidity```, that is the sum of ```fixed acidity``` and ```volatile acidity```.

````python
# Without renaming the columns we can use this line
# df['total_acidity'] = df['fixed acidity'] + df['volatile acidity']
# df.head(2)
df.rename(columns={'fixed acidity': 'fixed_acidity', 'volatile acidity': 'volatile_acidity'}, inplace=True)
df.eval('total_acidity = fixed_acidity + volatile_acidity', inplace=True)
df.head(4)
````

5. Find the average ```total_acidity``` for each of the ```quality values```.

````python
df.groupby('quality')['total_acidity'].mean()
````

6. Find the top 5 ```density``` values.

````python
df.sort_values('density',ascending=False).head(5)
df.sort_values('density',ascending=False)[:5]
````

7. Find the 10 lowest ```sulphates``` values.

````python
df.sort_values('sulphates',ascending=True)[:10]
df.sort_values('sulphates',ascending=True).head(10)
````

### Part 4: Perform exploratory analysis & visualization

1. If white wine quality 7 and above is considered good, what is the percentage of good quality wines produced.

````python
df[df.quality>=7].count()/df.count()*100
````

2. what is the alcohol quantity distribution among the "white wines"?

````python
import matplotlib.pyplot as plt
alchl_distbtn = df.loc[:,'alcohol']
plt.hist(alchl_distbtn,range=(8,14.2),bins=10,histtype='bar',color='green',label="Alcohol Distribution")
plt.xlabel("Alcohol distriution")
plt.ylabel("White wine")
plt.legend()
plt.show()
#print(bins)
#print(counts)
````
