```python
#1.Write a Python program to iterate over sets.
num_set = {1,4,4,5,6,9}
for n in num_set:
    print(n, end=' ')
print("\nCreating a set using string:")
string_set = {"sru","mah","ami","sru"}
for x in string_set:
    print(x, end =' ')
```

```python
#2.Write a Python program to add member(s) to a set.
num_set = {1,4,4,5,6,9}
num_set.add(99)
print(num_set)
print("\nAdding multiple elements")
num_set.update([77,90])
print(num_set)
```

```python
#3.Write a Python program to check if a set is a subset of another set.
x={"mango","apple"}
y={"apple"}

print(x)
print(y)

print("check if x is a subset of y")
print(x <= y)
print(x.issubset(y))

print("\ncheck if y is a subset of x")
print(y <= x)
print(y.issubset(x))
```

```python
#4 Write a Python program to create a shallow copy of sets.
set1 = set(["red","green"])
print(set1)
set2 = set1.copy()
print(set2)
```

```python
#5 Write a Python program to find the type,length ,maximum and minimum values in a set.

set1={20,99,90,56,87,13,5,83}
print(type(set1))
print("The length of the set", len(set1))
print("The maximum value in the set", max(set1))
print("The minimum value in the set", min(set1))
```

```python
#6 Write a Python program to check if a given value is present in a set or not.
set1={20,99,90,56,87,13,5,83}
print("check 56 exist in set1")
print(56 in set1)
print("check 6 not exist in set1")
print(6 not in set1)

```
