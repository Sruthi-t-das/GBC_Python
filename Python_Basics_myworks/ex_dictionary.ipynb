```python
#1.Write a Python program to sort (ascending and descending) a dictionary by value.
import operator
d = {1: 2, 3: 4, 4: 3, 2: 1, 0: 0}
print("The original dictionary", d)
sorted_asc = dict(sorted(d.items(), key = operator.itemgetter(1)))
print("The dictionary in ascending order by value", sorted_asc)
sorted_desc = dict(sorted(d.items(),key = operator.itemgetter(1),reverse = True))
print("The dictionary in descending order by value", sorted_desc)
```

```python
#2.Write a Python script to concatenate the following dictionaries to create a new one.
dic1 = {1: 10, 2: 20}
dic2 = {3: 30, 4: 40}
dic3 = {5: 50, 6: 60}
dic4 = {}
for d in (dic1,dic2,dic3):
    dic4.update(d)
print(dic4)
```

```python
#4.Write a Python program to check whether a given key already exists in a dictionary.

d = {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}
def is_keypresent(x):
    if x in d:
       print(f'The key {x} is present in dictionary')
    else:
       print(f'The key {x} is not present in dictionary')
is_keypresent(5)
is_keypresent(10)

```
