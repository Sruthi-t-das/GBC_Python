```python
# Write a program that prints all the prime numbers between 1 and 50 using a loop.
# For each number, check if it is prime and print it only if it is.
prime = []
not_prime = []
for x in range(1,51):
    if x<=1:
        not_prime.append(x)
    else:
        is_prime = True
        for i in range(2,int((x**0.5) +1)):
            if x%i == 0:
                is_prime = False
                break
        if is_prime:
            prime.append(x)
        else:
            not_prime.append(x)
    
print(prime)
print(not_prime)
```

Convert this list into a dictionary where:

Key = product name

Formula: quantity Ã— price

```python
inventory = [
    ("Laptop", 10, 1000),
    ("Mouse", 50, 25),
    ("Keyboard", 20, 50),
    ("Monitor", 15, 300)
]

# Step 1: Convert to dict {product: (qty, price)}
stock = {}
for name, qty, price in inventory:
    stock[name] = (qty, price)

# Step 2: Calculate total value per product
total_value = {}
for name, (qty, price) in stock.items():
    total_value[name] = qty * price

print(stock)
print(total_value)

```

```python
from itertools import combinations

dicts = {
    'math': {'Alice', 'Bob', 'David'},
    'english': {'Alice', 'Charlie', 'David'},
    'history': {'Alice', 'Charlie'},
    'physics': {'Bob'},
    'art': {'Charlie'}
}

result = []

# Step 1: Loop through all unique pairs of subjects
for sub1, sub2 in combinations(dicts.keys(), 2):
    # Step 2: Find common students using set intersection
    common_students = dicts[sub1].intersection(dicts[sub2])
    count = len(common_students)
    
    # Step 3: Add to result
    result.append((sub1, sub2, count))

# Print result
for item in result:
    print(f"{item[0]} & {item[1]} â†’ {item[2]} common students")
```
