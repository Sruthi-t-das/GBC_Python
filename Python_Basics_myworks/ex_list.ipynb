```python
# 1. Write a Python program to sum all the items in a list.
sample_list = [2,15,-90]
#define a function to find the sum of items in a list
def sum_list(items):     
    sum = 0                 #initialize sum to 0       
    for x in items:         #iterate through each item in a list
        sum += x            #add current item to the sum
    return sum              #return sum
#print the sum of items in the list
print("The sum is:", sum_list(sample_list))
```

```python
#2.Write a Python program to get the largest number from a list.

#define a function ti find the largest among the list
def max_list(items):
    max_val = items[0]           #Initialize max_val with the first item in the list
    for x in items:              #iterate each item in the list
        if x > max_val:          #check the current item greater than max_val
            max_val = x          #update max_val
    return max_val               #return the largest value found
#print the largest value among the list
print("The largest of the given list is", max_list([2,45,98,77]))
```

```python
#7.Write a Python program to check if each number is prime in a given list of numbers. Return True if all numbers are prime otherwise False.

def primecheck(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i == 0:
            return False
    return True




def allprime(num):
    primenums = [x for x in num if primecheck(x)]

    if primenums:
        print("prime numbers in the list", primenums)
    else:
        print("nothing is prime")
    
     

num_list = [12,55,67,33,42,68,7]
result = allprime(num_list)


```
