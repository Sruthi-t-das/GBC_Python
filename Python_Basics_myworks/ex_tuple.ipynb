```python
#1. Write a Python program to add an item to a tuple.

Tuple1=(5,3,8,9,90,45)
print("The tuple is",Tuple1)
Tuple1 = Tuple1 +(9,)
print("The tuple is",Tuple1)
Tuple1 = Tuple1[:5] + (23,57,88) + Tuple1[:5]
print("The tuple is",Tuple1)
```

```python
#2.Write a Python program to convert a tuple to a string.

tuple1 = ('s','r','u','t','h','i')
str = ''.join(tuple1)
print(str)
```

```python
#3.Write a Python program to get the 4th element from the last element of a tuple.
tuplex = ("w", 3, "r", "e", "s", "o", "u", "r", "c", "e")
print(tuplex)
print("\nThe third position item is:",tuplex[3])
print("\nThe -4th position item is:",tuplex[-4])

```

```python
#4.Write a Python program to find repeated items in a tuple.
tuplex = 2, 4, 5, 6, 2, 3, 4, 4, 7
print(tuplex)
count = tuplex.count(4)
print(count)
```

```python
#5.Write a Python program to remove an item from a tuple.

tuplex = "w", 3, "r", "s", "o", "u", "r", "c", "e"
print(tuplex)

tuplex = tuplex[:2] + tuplex[3:]
print(tuplex)

listx = list(tuplex)  # tuple is immutable, so we convert it to a list
listx.remove("c")     # remove "c" from the list
tuplex = tuple(listx) # convert back to tuple
print(tuplex)

```
